```python
import pandas as pd
import matplotlib.pyplot as plt
import emoji
```

```python
import re
import unicodedata

def remove_emoji(string):
    # Remove emojis using regex
    string = emoji.demojize(string)

    # Remove emojis from the text
    string = emoji.replace_emoji(string, '')
    string = re.sub(r':[a-z_]+:', '', string)
    

    return string
    
```

```python
df = pd.read_csv('shoppee.csv')
print(df.head(20))
```

```python
df['Price'] = [int(price/10000) for price in df['Price']]
df['Price']
```

```python
names = df['Name']
for name in names:
    name = remove_emoji(name)
df['Name'] =names
df['Name']
```

```python
df['Discount'].fillna('0%',inplace=True)
df
```

```python
df.to_csv('cleaned_data.csv',index=False)
```
